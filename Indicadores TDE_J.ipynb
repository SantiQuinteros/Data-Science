```python
indicators = {
    'NY.GDP.PCAP.CD': 'PIB per cápita (US$ a precios actuales)',

    'GB.XPD.RSDV.GD.ZS': 'Gasto en investigación y desarrollo (% del PIB)',
    'TX.VAL.TECH.MF.ZS': 'Exportaciones de productos de alta tecnología (% de las exportaciones de productos manufacturados)',
    'NY.GDP.PCAP.KD.ZG': 'Crecimiento del PIB per cápita (% anual)',
    'NY.GDP.DEFL.KD.ZG': 'Inflación, índice de deflación del PIB (% anual)',
    'IC.TAX.TOTL.CP.ZS': 'Tasa tributaria total (% de utilidades comerciales)',
    'FR.INR.RINR': 'Tasa de interés real (%)',

    'NE.IMP.GNFS.ZS': 'Importaciones de bienes y servicios (% del PIB)',
    'NE.EXP.GNFS.ZS': 'Exportaciones de bienes y servicios (% del PIB)',
    'TG.VAL.TOTL.GD.ZS': 'Comercio de mercaderías (% del PIB)',

    'NY.GNS.ICTR.ZS': 'Ahorro bruto (% del PIB)',
    'NV.AGR.TOTL.ZS': 'Agricultura, valor agregado (% del PIB)',
    'NE.GDI.TOTL.ZS': 'Formación bruta de capital (% del PIB)',
    'GC.XPN.TOTL.GD.ZS': 'Gasto (% del PIB)',
    'GC.DOD.TOTL.GD.ZS': 'Deuda del gobierno central, total (% del PIB)',

    'SE.XPD.TOTL.GD.ZS': 'Gasto público en educación, total (% del PIB)',
    'SE.PRM.ENRR': 'Inscripción escolar, nivel primario (% bruto)',
    'SE.SEC.ENRR': 'Inscripción escolar, nivel secundario (% bruto)',
    'SE.TER.ENRR': 'Inscripción escolar, nivel terciario (% bruto)',
    'SE.ADT.LITR.MA.ZS': 'Tasa de alfabetización, varones adultos (% de varones de 15 años o más)',
    'SI.DST.FRST.10': 'Participación en el ingreso del 10% peor remunerado de la población'


}
countries = ["KR", "UY", "PA", "NZ", "IL", "ZA", "DE"]
```

```python
data = wbdata.get_dataframe(indicators=indicators, country=countries)
```

```python
data.reset_index(inplace=True)
```

```python
data['date'] = pd.to_datetime(data['date'])
data['date'] = data['date'].dt.year
data['Exportaciones de productos de alta tecnología (% de las exportaciones de productos manufacturados)'] = data['Exportaciones de productos de alta tecnología (% de las exportaciones de productos manufacturados)'].where(data['Exportaciones de productos de alta tecnología (% de las exportaciones de productos manufacturados)'] <= 100, np.nan)
```

PIB PER CAPITA

```python
plt.style.use('seaborn-darkgrid')

grouped_data = data.groupby('country')

fig = plt.figure(figsize=(15, 10))

for country, group in grouped_data:
    plt.plot(group['date'], group['PIB per cápita (US$ a precios actuales)'], label=country)

plt.xlabel('Año', fontsize=20)
plt.ylabel('US$ a precios actuales', fontsize=20)
plt.title('PIB per cápita', fontsize=30)

ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))

ay = plt.gca()
ay.yaxis.set_major_locator(ticker.MultipleLocator(5000))

ax.tick_params(axis='x', labelsize=15)
ay.tick_params(axis='y', labelsize=15)

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, fontsize=20)

fig.patch.set_facecolor('lightgray')

plt.legend(fontsize=20)

plt.show()
```

```python
plt.style.use('seaborn-darkgrid')

countries = data['country'].unique()

fig, axes = plt.subplots(2, 3, figsize=(40, 20))

for i, country in enumerate(countries):
    ax = axes[i // 3, i % 3]
  
    country_data = data[data['country'] == country]
    
    ax.plot(country_data['date'], country_data['Importaciones de bienes y servicios (% del PIB)'], label='Importaciones')
    ax.plot(country_data['date'], country_data['Exportaciones de bienes y servicios (% del PIB)'], label='Exportaciones')
    ax.plot(country_data['date'], country_data['Comercio de mercaderías (% del PIB)'], label='Comercio')
    
    ax.set_xlabel('Año', fontsize=20)
    ax.set_ylabel('% del PIB', fontsize=20)
    ax.set_title(f'{country}', fontsize=30)
    
    ax.tick_params(axis='both', labelsize=20)
    
    ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
    
    ax.set_ylim(0, 100)
    
 
    ax.legend(fontsize=20)

fig.tight_layout()

plt.show()
```
